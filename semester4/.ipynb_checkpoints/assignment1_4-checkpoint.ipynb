```python
import numpy as np
import matplotlib.pyplot as plt

# Randomly selecting reaction rates from interval [0, 2]
k1, k2 = np.random.uniform(0, 2, 2)

# Randomly selecting initial concentrations from interval (0, 1) for A, B, C and setting D to 0
A0, B0, C0, D0 = np.random.uniform(0.01, 1, 4)

# Defining the system of differential equations
def reaction_system(t, y):
    A, B, C, D = y
    dA_dt = -k1 * A
    dB_dt = k1 * A - k2 * B * C
    dC_dt = k1 * A - k2 * B * C
    dD_dt = k2 * B * C
    return [dA_dt, dB_dt, dC_dt, dD_dt]

# Time points to solve the system
t = np.linspace(0, 10, 100)  # 10 time units, 100 points

# Runge-Kutta 4th Order Method (RK4) to solve the system
from scipy.integrate import solve_ivp

solution = solve_ivp(reaction_system, [t[0], t[-1]], [A0, B0, C0, D0], t_eval=t)

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(solution.t, solution.y[0], label='A')
plt.plot(solution.t, solution.y[1], label='B')
plt.plot(solution.t, solution.y[2], label='C')
plt.plot(solution.t, solution.y[3], label='D')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.title('Concentration vs Time for Reactions')
plt.legend()
plt.grid(True)
plt.show()

k1, k2, (A0, B0, C0, D0)  # Displaying the selected reaction rates and initial conditions
```

```python
import numpy as np
import matplotlib.pyplot as plt

def reaction_rates(y):
    A, B, C, D = y
    dAdt = -rGreitis1 * A
    dBdt = rGreitis1/2 * A - rGreitis2 * B * C
    dCdt = rGreitis1/2 * A - rGreitis2 * C * B
    dDdt = rGreitis2 * C * B
    return np.array([dAdt, dBdt, dCdt, dDdt])

def rk4_step(xn, yn, dx):
    k1 = dx * reaction_rates(yn)
    k2 = dx * reaction_rates(yn + 0.5*k1)
    k3 = dx * reaction_rates(yn + 0.5*k2)
    k4 = dx * reaction_rates(yn + k3)
    yn1 = yn + (k1 + 2*k2 + 2*k3 + k4) / 6.0
    xn1 = xn + dx
    return xn1, yn1

rGreitis1 = 2
rGreitis2 = 2

A0 = 1
B0 = 0.05
C0 = 0.05
D0 = 0.02
y0 = np.array([A0, B0, C0, D0])

a = 0
b = 20
N = 200
dx = (b - a) / N
t = np.linspace(a, b, N)
results = []

y = y0
for tn in t:
    results.append(y)
    tn1, y = rk4_step(tn, y, dx)

results = np.array(results)

plt.plot(t, results[:, 0], label='A(t)')
plt.plot(t, results[:, 1], label='B(t)')
plt.plot(t, results[:, 2], label='C(t)')
plt.plot(t, results[:, 3], label='D(t)')
plt.xlabel('t, laikas')
plt.ylabel('Koncentracija')
plt.title('Kitimas')
plt.legend()
plt.show()
```
