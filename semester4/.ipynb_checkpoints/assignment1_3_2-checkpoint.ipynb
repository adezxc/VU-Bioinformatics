```python
import numpy as np
import matplotlib.pyplot as plt

def generate_data(sigma, T):
    t = np.linspace(0, T, 100)
    eps = np.random.normal(0, sigma, 100)
    for i in range(100):
        a, b, c = np.random.rand(3) * (1 - (-1)) + (-1)
    x = a * t**2 + b * t + c + eps
    x_without_eps = a * t**2 + b * t + c
    plt.plot(x_without_eps, 'bo')
    
    return x

def calculate_error(data, a, b, c, T):
    t = np.linspace(0, T, 100)
    predicted_data = a * t**2 + b * t + c
    error = np.sum((data - predicted_data)**2)
    return error

def find_best_parameters(initial_data, T):
    best_error = float('inf')
    best_parameters = None

    for a in np.arange(-1, 1, 0.05):
        for b in np.arange(-1, 1, 0.05):
            for c in np.arange(-1, 1, 0.05):
                error = calculate_error(initial_data, a, b, c, T)
                if error < best_error:
                    best_error = error
                    best_parameters = (a, b, c)

    return best_parameters

T = 1
t = np.linspace(0, T, 100)
initial_data = generate_data(0.2, T)
plt.plot(initial_data, 'bo')
best_parameters = find_best_parameters(initial_data, T)
print(best_parameters)
x_with_best_params = best_parameters[0] * t**2 + best_parameters[1] * t + best_parameters[2]
plt.plot(x_with_best_params, color = 'red')

```
