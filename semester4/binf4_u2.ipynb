```
# 2012 % 12 + 1 = 9
# Variantas 9

import sympy as sp
import matplotlib.pyplot as plt

# Define symbolic variables
x, alpha = sp.symbols('x alpha')

f = sp.cos(-alpha * x + sp.pi/6)

trapezoid_results = {}
simpson_results = {}

for a in [a/100  for a in range(1, 101, 1)]:
    # f(a)
    fa = f.subs({alpha: a, x: 0})
    # f(b)
    fb = f.subs({alpha:a, x: 1})

    trapezoid_result = (1 - 0) * 1/2 * (fa + fb)
    simpson_result = ((1 - 0) / 6) * (fa + 4 * f.subs({alpha: a, x: (1 + 0 / 2)}) + fb)

    trapezoid_results[a] = trapezoid_result
    simpson_results[a] = simpson_result

x_values = list(trapezoid_results.keys())
y_values = [float(result) for result in trapezoid_results.values()]

x_values2 = list(simpson_results.keys())
y_values2 = [float(result) for result in simpson_results.values()]

plt.plot(x_values, y_values, label='Trapezoidal Rule')
plt.plot(x_values2, y_values2, label="Simpson's Rule")
plt.xlabel('α')
plt.ylabel('Integralo reikšmė')
plt.title('Integralo reikšmės keičiant α')
plt.legend()
plt.show()
```
