```python
from ete3 import Tree, TreeStyle, NodeStyle, AttrFace, TextFace

class SuffixTree:
    def __init__(self, string):
        self.root = {}
        self.build_suffix_tree(string)

    def build_suffix_tree(self, string):
        n = len(string)
        for i in range(n):
            current = self.root
            for j in range(i, n):
                character = string[j]
                if character not in current:
                    current[character] = {}
                current = current[character]

    def _build_ete_tree(self, current_node, ete_node):
        for char, child_node in current_node.items():
            new_ete_node = ete_node.add_child(name=char)
            self._build_ete_tree(child_node, new_ete_node)

    from ete3 import Tree, TreeStyle, NodeStyle, AttrFace, TextFace

class SuffixTree:
    def __init__(self, string):
        self.root = {}
        self.build_suffix_tree(string)

    def build_suffix_tree(self, string):
        n = len(string)
        for i in range(n):
            current = self.root
            for j in range(i, n):
                character = string[j]
                if character not in current:
                    current[character] = {}
                current = current[character]

    def _build_ete_tree(self, current_node, ete_node):
        for char, child_node in current_node.items():
            new_ete_node = ete_node.add_child(name=char)
            self._build_ete_tree(child_node, new_ete_node)

    def visualize(self):
        ete_tree = Tree()
        self._build_ete_tree(self.root, ete_tree)

        ts = TreeStyle()
        ts.show_leaf_name = False
        ts.show_branch_length = False
        ts.scale = 30

        nstyle = NodeStyle()
        nstyle["shape"] = "circle"
        nstyle["fgcolor"] = "black"
        nstyle["size"] = 10
        nstyle["bgcolor"] = "white"

        for n in ete_tree.traverse():
            n.set_style(nstyle)
            char_face = TextFace(n.name, tight_text=True)
            n.add_face(char_face, column=0, position="branch-right")

        ete_tree.show(tree_style=ts)

# Example usage
string = "bananaasdf$"
tree = SuffixTree(string)
tree.visualize()


# Example usage
string = "bananaasdf$"
tree = SuffixTree(string)
tree.visualize()
```
