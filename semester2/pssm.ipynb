```python
from Bio import AlignIO
from Bio import SeqIO
from Bio import Seq
import os

sequences = []
ids = []
with open("example.fasta") as handle:
    for record in SeqIO.parse(handle, "fasta"):
        sequences.append(str(record.seq))
        ids.append(str(record.id))
handle.close()

print(sequences)
print(ids)
```

```python
#count occurrences

A = []
G = []
C = []
T = []
for s in zip(*sequences):
    print(s)
    num_a = 0
    num_g = 0
    num_c = 0
    num_t = 0
    for nuc in s:
        if nuc == "A":
            num_a += 1
        elif nuc == "G":
            num_g += 1
        elif nuc == "C":
            num_c += 1
        elif nuc == "T":
            num_t += 1
    A.append(num_a)
    G.append(num_g)
    C.append(num_c)
    T.append(num_t)


print(A, G, C, T)
```

```python
import pandas as pd
import numpy as np

table_count = {'A': A, 'G': G, 'C': C, 'T': T}
table_draw = pd.DataFrame(table_count)
table_draw.index += 1
table_draw_tr = table_draw.transpose()
table_draw_tr
```

```python
#count frequances
A_freq = []
G_freq = []
C_freq = []
T_freq = []

how_many_seq = len(ids)
for i in A:
    freq = i/how_many_seq
    A_freq.append(freq)

for i in G:
    freq = i/how_many_seq
    G_freq.append(freq)

for i in C:
    freq = i/how_many_seq
    C_freq.append(freq)  

for i in T:
    freq = i/how_many_seq
    T_freq.append(freq)

table_freq = {'A': A_freq, 'G': G_freq, 'C': C_freq, 'T': T_freq}
table_freq_draw = pd.DataFrame(table_freq)
table_freq_draw.index += 1
table_freq_draw_tr = table_freq_draw.transpose()
table_freq_draw_tr
```

```python
#normalise values
overall_freq_A = np.mean(A_freq)
overall_freq_G = np.mean(G_freq)
overall_freq_C = np.mean(C_freq)
overall_freq_T = np.mean(T_freq)

#print (overall_freq_A, overall_freq_G, overall_freq_C, overall_freq_T)

A_log = []
G_log = []
C_log = []
T_log = []

for i in A_freq:
    normalised = i/overall_freq_A
    if normalised == 0:
        A_log.append(0)
        continue
    A_log.append(np.log2(normalised))

for i in G_freq:
    normalised = i/overall_freq_G
    if normalised == 0:
        G_log.append(0)
        continue
    G_log.append(normalised)

for i in C_freq:
    
    normalised = i/overall_freq_C
    if normalised == 0:
        C_log.append(0)
        continue
    C_log.append(normalised) 

for i in T_freq:
    normalised = i/overall_freq_T
    if normalised == 0:
        T_log.append(0)
        continue
    T_log.append(normalised)

table_norm = {'A': A_log, 'G': G_log, 'C': C_log, 'T': T_log}
table_norm_draw = pd.DataFrame(table_norm)
table_norm_draw.index += 1
table_norm_draw_tr = table_norm_draw.transpose()
table_norm_draw_tr
```
